```
from tqdm import tqdm
import csv

def makeFramePathFile(img_dir, output_file):
  # check if file exists and erases it if it does
  if os.path.isfile(output_file):
    print("\nOld file being deleted")
    os.system("rm "+output_file)

  count = 0
  with open(output_file, 'w') as f:
    for entry in tqdm(os.listdir(img_dir)):
      if entry == output_file:
        pass
      else:
        f.write(os.path.join(img_dir, entry)+"\n")
        count += 1
  
  print("Number of files in directory: "+str(count))
  
```

```
# make train image path file
os.chdir("/content/drive/MyDrive/Perception/INDY_DATASETS/alonsos_lightning_lap/images/train")
img_dir = os.getcwd()
makeFramePathFile(img_dir, 'image_paths.train')

# make val image path file
os.chdir("/content/drive/MyDrive/Perception/INDY_DATASETS/alonsos_lightning_lap/images/val")
img_dir = os.getcwd()
makeFramePathFile(img_dir, 'image_paths.val')
```
