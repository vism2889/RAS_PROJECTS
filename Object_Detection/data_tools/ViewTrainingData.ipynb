```
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg
from tqdm import tqdm

img_dir = os.getcwd()
print(img_dir)
count = 0
save_dir = "/content/drive/.shortcut-targets-by-id/1mQtFwgJglT1W1ne098FCmy-KLCv-7Gsd/Perception/INDY_DATASETS/alonsos_lightning_lap/train_with_labels/"

if not os.path.exists(save_dir):
  os.makedirs(save_dir)
else:
  os.system("rm -r "+save_dir)
  os.makedirs(save_dir)

for entry in tqdm(os.listdir(img_dir)):
  if entry == "image_paths.train":
    pass
  else:
    fig, ax = plt.subplots()
    img = mpimg.imread(entry)
    ax.imshow(img)

    img_width = 1280
    img_height = 720

    ann_file = entry.replace('jpg', 'txt')
    ann_path = "/content/drive/.shortcut-targets-by-id/1mQtFwgJglT1W1ne098FCmy-KLCv-7Gsd/Perception/INDY_DATASETS/alonsos_lightning_lap/labels_with_ids/train/"+ann_file
    f = open(ann_path, "r")
    lines = f.readlines()
    for line in lines:
      new_line = line.split()
      x_center = float(new_line[2])*img_width
      y_center = float(new_line[3])*img_height
      width = float(new_line[4])*img_width
      height = float(new_line[5])*img_height

      x_lower_left = x_center - float(width/2)
      y_lower_left = y_center - float(height/2)

      rect = patches.Rectangle((x_lower_left, y_lower_left), width, height, linewidth=2, edgecolor="r", fill=0)
      ax.add_patch(rect)
    
    fig.savefig(save_dir+entry)
    plt.close(fig)

    if count != 1:
      plt.show()
      count = 1
  
```
